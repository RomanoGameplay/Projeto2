<h1>Plataformas</h1>

```python
df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv', encoding='latin-1')
df_plat = df.copy()
df_plat.head()
```

```python
df_plat.rename(columns={'Name':'Game', 'Year_of_Release':'Year', 'NA_Sales':'North America', 'EU_Sales':'Europe', 'JP_Sales':'Japan', 'Other_Sales':'Rest of World', 'Global_Sales':'Global'}, inplace=True)
df_plat.head()
```

```python
df_plat.describe()
```

```python
df_plat.describe(include=object)
```

```python
df_plat.isnull().sum()
```

```python
df_plat.dropna(subset='Year', inplace=True)
df_plat.dropna(subset='Game', inplace=True)
df_plat.dropna(subset='Critic_Score', inplace=True)

mediana = df_plat['User_Score'].median()
df_plat.User_Score.fillna(mediana, inplace=True)

mediana = int(df_plat['User_Count'].median())
df_plat.User_Count.fillna(mediana, inplace=True)

df_plat.Publisher.fillna('-', inplace=True)
df_plat.isnull().sum()
```

```python
duplicates(df_plat, ['Game', 'Platform'])
```

```python
df_plat.loc[(df_plat['Game'] == 'Need for Speed: Most Wanted') & (df_plat['Year'] == 2005.0) &  (df_plat['Platform'] == 'X360'), 'Game'] = 'Need for Speed: Most Wanted (2005)'
df_plat.loc[(df_plat['Game'] == 'Need for Speed: Most Wanted') & (df_plat['Year'] == 2012.0) &  (df_plat['Platform'] == 'X360'), 'Game'] = 'Need for Speed: Most Wanted (2012)'

df_plat.loc[(df_plat['Game'] == 'Need for Speed: Most Wanted') & (df_plat['Year'] == 2005.0) &  (df_plat['Platform'] == 'PC'), 'Game'] = 'Need for Speed: Most Wanted (2005)'
df_plat.loc[(df_plat['Game'] == 'Need for Speed: Most Wanted') & (df_plat['Year'] == 2012.0) &  (df_plat['Platform'] == 'PC'), 'Game'] = 'Need for Speed: Most Wanted (2012)'
```

```python
duplicates(df_plat, ['Game', 'Platform'])
```

```python
df_plat.drop_duplicates(subset=['Game', 'Platform'], inplace=True, keep='first')
```

```python
df_plat['Global'] = df_plat.loc[:, ['North America', 'Europe', 'Japan', 'Rest of World']].sum(axis=1)
df_plat.head()
```

```python
df_plat.isnull().sum()
```

```python
df_plat.Developer.fillna('-', inplace=True)
df_plat.fillna('RP', inplace=True)
df_plat.isnull().sum()
```

```python
df_plat.Rating.replace('K-A', 'E', inplace=True)
df_plat.Rating.value_counts()
```

```python
df_plat['Year'] = pd.to_datetime(df_plat['Year'], format='%Y')
df_plat['Year'] = df_plat.Year.dt.year
df_plat.head()
```
